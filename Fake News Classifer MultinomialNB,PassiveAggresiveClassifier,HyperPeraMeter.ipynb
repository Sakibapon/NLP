```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
x=df.drop('label',axis=1)
y=df['label']
df.shape
```

```python
df=df.dropna()
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer,HashingVectorizer
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
ps=PorterStemmer()
lm=WordNetLemmatizer()
corpus=[]
```

```python
msg=df.copy()
msg.reset_index(inplace=True)
msg.head()
```

```python
msg['title'][1]
```

```python
for i in range(len(msg)):
    words=re.sub('[^a-zA-z]',' ',msg['title'][i])
    words=words.lower()
    words=words.split()
    words=[ps.stem(word)for word in words if word not in stopwords.words('english')]
    words=' '.join(words)
    corpus.append(words)
```

```python
cv=CountVectorizer(max_features=6000,ngram_range=(1,3))
x=cv.fit_transform(corpus).toarray()
```

```python
y=msg['label']
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
cv.get_feature_names()
```

```python
cv.get_params()
```

```python
pdf=pd.DataFrame(X_train, columns=cv.get_feature_names())
pdf
```

# Multinomial  NB

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report,accuracy_score
classifier=MultinomialNB()
classifier.fit(X_train,y_train)
pred=classifier.predict(X_test)
score=accuracy_score(y_test, pred)
score
```

```python
print(classification_report(y_test,pred))
```

# Passive Aggresive Classifier

```python
import itertools
from sklearn.linear_model import PassiveAggressiveClassifier
pac=PassiveAggressiveClassifier()
pac.fit(X_train,y_train)
pred=pac.predict(X_test)
score=accuracy_score(y_test,pred)
print(score)
print(classification_report(y_test,pred))
```

# MultiNomial NB with HyperParaMeter

```python
cs=MultinomialNB(alpha=0.1)
bscore=0
```

```python
for alpha in np.arange(0,1,0.1):
    subcs=MultinomialNB(alpha=alpha)
    subcs.fit(X_train,y_train)
    predi=subcs.predict(X_test)
    score=accuracy_score(y_test,predi)
    if score>bscore:
        cs=subcs
        print('Alpha Rate= ',alpha,'And Score= ',score)
```
