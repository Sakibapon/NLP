```
import random
import re


# dictionary
transform = {"am": "are",
             "was": "were",
             "I": "you",
             "I would": "you would",
             "I have": "you have",
             "I will": "you will",
             "are": "am",
             "your": "my",
             "yours": "mine",
             "you": "me",
             "me": "you",
             "my": "your"}

answer = {'I need': ["why do you need", "Are you sure you need"],
            'I would': ["could you explain why you would", "why would you"],
            'crave': ["Why don't you tell me more about your cravings"],
            'I want': ["What would it mean to you if you got", "Why do you want", "What would you do if you got"],
            'eat': ["Why are you hungry?"],
            'what': ["Why do you ask", "How would an answer to that help you", "What do you think"],
            'anyq': ["Why do you ask that", "Please consider whether you can answer your own question.", "Perhaps the answer lies within yourself", "Why don't you tell me"],
            'anyem': ["Please tell me more.", "Can you elaborate on that?", "Very interesting.", "How does that make you feel?"],
            'because': ["Is that the real reason?", "Does that reason apply to anything else"],
            'course': ["This is CSE431- Natural Language Processing"],
            'faculty' : []}
             

# The chat takes string input from the user. 
chat = input("Hello there, What's your name? \n")

#the name from the chat.

if not re.findall(r'([Bb]ye)$|([Ee]xit)$|([Qu]it)$', chat):
  name = re.findall(r'\b[A-Za-z][a-z]+.?$', chat)
 
  print("Hi {}, how can I help you today?\n".format(
      " ".join(name).replace(".", "")))
else:
  pass




while not re.findall(r'([Bb]ye)$|([Ee]xit)$|([Qu]it)$', chat):
  chat = input()
  #input check 
  if re.findall(r'I need (.*)', chat):
    for k, ask in answer.items():
      if 'I need' not in k:
        continue
      else:
        tran_key = transform.keys()

        ahead = chat.split(r"need", 1)[1]
        newstring = ahead.split()
        for i in range(0, len(newstring)):
          if newstring[i] in tran_key:
            newstring[i] = transform[newstring[i]]
        print("".join(name) + ", " + random.choice(ask) +
              ' {}?'.format(" ".join(newstring)))

  
  elif re.findall(r'I want (.*)', chat):
    for k, ask in answer.items():
      if 'I want' not in k:
        continue
      else:
        tran_key = transform .keys()

        ahead = chat.split(r"want", 1)[1]
        newstring = ahead.split()
        for i in range(0, len(newstring)):
          if newstring[i] in tran_key:
            newstring[i] = transform[newstring[i]]
        print("".join(name) + ", " + random.choice(ask) +
              ' {}?'.format(" ".join(newstring)))

  
  elif re.findall(r'I would (.*)', chat):
    for k, ask in answer.items():
      if 'I would' not in k:
        continue
      else:
        tran_key = transform .keys()

        ahead = chat.split(r"would", 1)[1]
        newstring = ahead.split()
        for i in range(0, len(newstring)):
          if newstring[i] in tran_key:
            newstring[i] = transform [newstring[i]]
        print("".join(name) + ", " + random.choice(ask) +
              ' {}?'.format(" ".join(newstring)))

  elif re.findall(r'(.*)[Be]cause(.*)', chat):
    for k, ask in answer.items():
      if 'because' not in k:
        continue
      else:
        print("".join(name) + ", " + random.choice(ask) + '?')

  elif re.findall(r'(.*)[Cr]ave(.*)', chat):  
    for k, ask in answer.items():
      if 'crave' not in k:
        continue
      else:
        print("".join(name) + ", " + random.choice(ask) + '?')

  elif re.findall(r'(.*)[Ee]at(.*)', chat):  
    for k, ask in answer.items():
      if 'eat' not in k:
        continue
      else:
        print("".join(name) + ", " + random.choice(ask)+ "?")
        
  elif re.findall(r'(.*)[Cc]ourse(.*)', chat):  
    for k, ask in answer.items():
      if 'course' not in k:
        continue
      else:
        print("".join(name) + ", " + random.choice(ask)+ ", anything else ?")
        
  elif re.findall(r'(.*)[Ff]aculty(.*)', chat):  
      if 'faculty' not in k:
        continue
      else:
        print("".join(name) + ", " + "Mr. Annajiat Alim Rasel")
        

  elif re.findall(r'(.*)[Wh]ats?', chat):  # spotting 
    for k, ask in answer.items():
      if 'what' not in k:
        continue
      else:
        print("".join(name) + ", " + random.choice(ask) + '?')

  elif re.findall(r'(.*)[Be]cause', chat):
    for k, ask in answer.items():
      if 'because' not in k:
        continue
      else:
        print("".join(name) + ", " + random.choice(ask) + '?')

  elif re.findall(r'(.*)\?', chat):
    for k, ask in answer.items():
      if 'anyq' not in k:
        continue
      else:
        print("".join(name) + ", " + random.choice(ask) + '?')

  elif re.findall(r'([Bb]ye)$|([Ee]xit)$|([Qu]it)$', chat):
    break

 
  elif re.findall(r'(.*)', chat):
    for k, ask in answer.items():
      if 'anyem' not in k:
        continue
      else:
        print("".join(name) + ", " + random.choice(ask))

  else:
    pass

print("bye! see you soon.")

```
